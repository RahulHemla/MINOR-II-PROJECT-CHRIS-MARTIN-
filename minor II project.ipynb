#####MINOR II PROJECT (CHRIS MARTIN)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
ascii_chars = np.array(list("@%#*+=-:. "))
```

```python
image =plt.imread("chris_martin.jpg")
```

```python
if image.ndim==3:image = image.mean(axis=2)
```

```python
new_width=40
new_height=20
```

```python
height,width=image.shape
x_idx=np.linspace(0,width-1,new_width).astype(int)
y_idx=np.linspace(0,height-1,new_height).astype(int)
image=image[y_idx][:,x_idx]
```

```python
image=image/image.max()
```

```python
ascii_image=ascii_chars[(image*(len(ascii_chars)-1)).astype(int)]
```

```python
for row in ascii_image:
    print("".join(row))
```
